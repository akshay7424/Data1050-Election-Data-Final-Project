```python
import pandas as pd
import geopandas as gpd
import geoplot as gplt
import geoplot.crs as gcrs # (projections)
# %matplotlib inline
```

```python
from collections import Counter
state_counter = Counter()
city_counter = Counter()
```

```python
for state in df.state_code:
    state_counter[state] += 1
for city in df.city:
    city_counter[city] += 1
print(state_counter)
print()
print(city_counter)
```

```python
contiguous_usa = gpd.read_file(
  gplt.datasets.get_path('contiguous_usa')
)
contiguous_usa = contiguous_usa.sort_values('state').reset_index(drop=True) # Alphabetize the states list
contiguous_usa.head(3)
```

```python
gplt.polyplot(contiguous_usa, projection=gcrs.AlbersEqualArea())
```

```python
url = ('https://raw.githubusercontent.com/'
       'jasonong/List-of-US-States/master/states.csv')
abbreviations = pd.read_csv(url)
abbreviations.head(3)
```

```python
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.longitude, df.latitude))
zip_code_5 = [s[:5] for s in list(gdf.zip_code)]
```

```python
zipcodes = pd.read_csv('free-zipcode-database-Primary.csv')
zipcodes = zipcodes[['Zipcode','Lat','Long','ZipCodeType','City','State']]

startup_states = list(state_counter.keys())
zipcodes = zipcodes.query('State in @startup_states')
zipcodes = gpd.GeoDataFrame(zipcodes)
zipcodes.Zipcode
```

```python
z = list(zipcodes.Zipcode)
zipcodes = zipcodes.assign(Zipcode_string = [str(z[i]) for i in range(len(z))])
zipcodes
```

```python
gdf_merged = gdf.merge(zipcodes, left_on='zip_code', right_on='Zipcode_string')
gdf_merged
```

```python
# gdf_merged.assign(lat_diff = gdf_merged.latitude - gdf_merged.Lat)
```

```python
ax = gplt.polyplot(
    contiguous_usa,
    projection=gcrs.AlbersEqualArea()
)
gplt.pointplot(gdf, ax=ax)
```

```python
gdf2 = (contiguous_usa
      .merge(abbreviations, left_on='state', right_on='State')
      .merge(gdf_merged, left_on='Abbreviation', right_on='state_code'))

gdf2.head(2)
```

```python
gdf3 = gdf2[['state','geometry_x','status_closed']]
gdf3
```

```python
gdf4 = gdf3.groupby('state').agg('mean').reset_index().merge(contiguous_usa, left_on='state', right_on='state')
gdf4.head()
```

```python
gdf5 = gdf4[['state','status_closed','geometry']]
gdf5.head()
```

```python
gdf6 = gpd.GeoDataFrame(gdf5).assign(status_acquired = 1 - gdf5.status_closed)
gdf6.head()
```

```python
gplt.choropleth(
    gdf6,
    hue='status_acquired',
    cmap='RdYlGn',
    edgecolor='Black',
    legend=True,
    projection=gcrs.AlbersEqualArea(),
    figsize=(20, 8)
)
```

```python
ax = gplt.polyplot(
    contiguous_usa,
    projection=gcrs.AlbersEqualArea()
)

gplt.pointplot(gdf,
               ax=ax,
               hue='status_closed')
```

```python
# Drop the outlier rows 220, 257, 432, and 906:
gdf_trimmed = gdf.drop([220,257,432,906])
gdf_trimmed2 = gdf_trimmed.assign(status_acquired = 1 - gdf_trimmed.status_closed)
```

```python
ax = gplt.polyplot(
    contiguous_usa,
    projection=gcrs.AlbersEqualArea(),
    figsize=(12,12)
)
gplt.pointplot(gdf_trimmed2,
               ax=ax,
              hue='status_acquired',
              cmap='RdYlGn',
              scale='status_acquired', limits=(4,4))
```
